# Data Cleaning

## Enron's Emails

1) I need to obtain the first names and last names of the individuals in 'From' and 'To' columns to remove them from the contents of the email.
<br>
<br>
2) The reason is I do not want the generator generating names

```python
enron_df
```

```python
enron_df[['From', 'To']] = enron_df[['From', 'To']].replace("frozenset\({'", "", regex=True)
enron_df[['From', 'To']] = enron_df[['From', 'To']].replace(r"\'}\)", "", regex=True)
enron_df[['From', 'To']] = enron_df[['From', 'To']].replace(r"\'", "", regex=True)
enron_df[['From', 'To']] = enron_df[['From', 'To']].replace("\@\S+", "", regex=True)
enron_df[['From', 'To']] = enron_df[['From', 'To']].replace("\.", " ", regex=True)
enron_df
```

<br>
I defined a function to add the first and last names of a column to a list. In addition, I have removed duplicates from the list to make it as short as possible.
<br>
<br>

```python
def gather_stopwords(df):
    my_list = []
    for i in df:
        for j in range(len(str(i).split())):
            my_list.append(str(i).split()[j])
            
    my_list = sorted(list(dict.fromkeys(my_list)), key=len)
    return my_list
```

```python
stopwords_from = gather_stopwords(enron_df['From'])
stopwords_to = gather_stopwords(enron_df['To'])
stopwords_combined = list(dict.fromkeys(stopwords_from + stopwords_to))
stopwords_combined = sorted([i.capitalize() for i in stopwords_combined], key=len)
```

```python
len(stopwords_combined)
```

```python
stopwords_combined
```

<br>
1) The list contains more than names, it contains symbols and some english words such as 'I' and 'Do'
<br>
<br>
2) I will remove words with less than 3 letters.
<br>
<br>

```python
stopwords_combined = [i for i in stopwords_combined if len(i) > 2]
stopwords_combined
```

```python
len(stopwords_combined)
```

```python
enron_df = enron_df.drop(labels=['From', 'To'], axis=1)
enron_df
```

<br>
1) This is much better, we now have the first and last names in the list.
<br>
<br>
2) I no longer need the 'From' and 'To' columns in Enron's dataframe and droped them.
<br>

```python
pd.set_option('display.max_colwidth', None)
enron_df.head(10)
```

## Comments
<br>
1) There are some instances of '= ' found in the middle of the word and between words. Examples are 'exp= ected', 'la= ck' and 'other = aspects'.
<br>
<br>
2) There are also alot of serial numbers (e.g. APRS0119234206) and phone numbers (e.g. 510-987-9193). These must be cleaned.
<br>
<br>
3) Forwarded emails have a line of information detailing 'From', 'Date', 'Time' and 'To'
<br>
<br>
4) There are special characters, e.g. '-', '_'

```python
# Issue 1
enron_df['content'] = enron_df['content'].replace('=', ' ', regex=True)
# Issue 2
enron_df['content'] = enron_df['content'].replace(r'\d+',' ', regex=True)
# Issue 3
enron_df['content'] = enron_df['content'].replace('---------------------- Forwarded by(.*)Subject:',\
                                                 ' ', regex=True)
# Issue 4
enron_df['content'] = enron_df['content'].replace(to_replace=r'[^\w\s]|_', \
                                                  value= ' ', \
                                                  regex=True)
```

```python
enron_df.head(20)
```

<br>
Removing words in the list 'stopwords_combined' from the 'Subject' and 'content' columns

```python
pd.set_option('display.max_colwidth', None)
enron_df['Subject'] = enron_df['Subject'].astype(str)
enron_df['content'] = enron_df['content'].astype(str)

enron_df['Subject'] = [' '.join([i for i in x.split() if i not in stopwords_combined]) for x in \
                       enron_df['Subject']]
enron_df['content'] = [' '.join([i for i in x.split() if i not in stopwords_combined]) for x in \
                       enron_df['content']]

enron_df = enron_df.rename(columns={'content':'Content'})
enron_df.head()
```

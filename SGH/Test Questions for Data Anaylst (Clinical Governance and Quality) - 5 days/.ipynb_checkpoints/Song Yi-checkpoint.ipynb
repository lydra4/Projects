# Merging Dataframe

```python
malaria_deaths
```

```python
malaria_inc
```

```python
malaria_deaths[malaria_deaths['Entity'] == 'Afghanistan']
```

```python
malaria_inc[malaria_inc['Entity'] == 'Afghanistan']
```

<br>
1) Malaria Deaths has data from 1990 up to 2016. However Malaria Incidence has only data from the years 2000, 2005, 2010 and 2015.
<br>
<br>
2) This means that after merging the 2 mentioned dataframes, there will be missing values.
<br>
<br>

```python
malaria_deaths_inc = malaria_deaths.merge(right=malaria_inc.iloc[:,[0,2,3] ], \
                                          how='left', \
                                          on=['Entity', 'Year'])
malaria_deaths_inc.head(26)
```

```python
malaria_deaths_inc = malaria_deaths_inc.rename(columns={malaria_deaths_inc.columns[3]: 'Deaths'})
malaria_deaths_inc = malaria_deaths_inc.rename(columns={malaria_deaths_inc.columns[4]: 'Incidence'})
malaria_deaths_inc.head(26)
```

```python
fig = px.scatter(data_frame=malaria_deaths_inc,
           x='Year', \
           y='Deaths', \
          color='Entity')

fig.update_layout(title={
                   'text': 'Death Rate by Country',
                    'x':0.5,
                    'y':0.95,
                    'xanchor': 'center',
                    'yanchor': 'top'
})

fig.show();
```

```python
fig = px.line(data_frame=malaria_deaths_inc.groupby('Year').mean().fillna(value=''),\
              x=malaria_deaths_inc.groupby('Year').mean().fillna(value='').index,\
              y = malaria_deaths_inc.groupby('Year').mean().fillna(value='')['Incidence'])

fig.update_traces(connectgaps=True)

fig.update_layout(title={
                   'text': 'Incidence Rate by Country',
                    'x':0.5,
                    'y':0.88,
                    'xanchor': 'center',
                    'yanchor': 'top'
},
                 xaxis_range=[1999,2016],
                 xaxis_title='Year',
                 yaxis_title='Incidence')
```

## Comments
<br>
1) To strength the view that incidence and mortality rate has a negative correlation with the year, the above 2 charts depict the decline in the above 2 statistics.
<br>
<br>
2) The death rate has also declined across all countries.
<br>
<br>

```python
malaria_deaths_age = malaria_deaths_age.iloc[:, 1:]
malaria_deaths_age
```

```python
fig = px.line(data_frame=malaria_deaths_age.groupby(['age_group', 'year']).mean().reset_index(),\
              x='year', y='deaths', color='age_group')

fig.update_layout(title={
    'text': 'Trended Death Rate by Age',
    'x': 0.5,
    'y': 0.95
})
```

## Comments
<br>
1) The decline in overall death rate due to Malaria is due a immense fall in death rate for those under 5 years old.
<br>
<br>
2) Even though the other age groups have recorded a rise in the death rate, the decline in death rate for those below 5 y.o. has more than offset the increase in other age groups.
<br>
<br>
3) This has resulted in an overall decline in Malaria's mortality rate.
